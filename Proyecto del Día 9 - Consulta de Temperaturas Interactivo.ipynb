```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ruta = 'C:/Users/KeythiCL/Downloads/Datos+Meteorológicos_Arg_2023 (1).csv'
df = pd.read_csv(ruta)
df
```

```python
type(df['Fecha'][0])
```

```python
df['Fecha'] = pd.to_datetime(df['Fecha'], format = '%d/%m/%Y')
df
```

```python
df['Fecha'][0]
```

```python
ciudades_unicas = df['Ciudad'].drop_duplicates().tolist()

# Mostrar las ciudades únicas al usuario
print(f'Ciudades disponibles: {ciudades_unicas}')

# Permitir al usuario seleccionar una ciudad
ciudad_seleccionada = input('Por favor, seleccione una ciudad de la lista: ')
mes_seleccion = int(input('Por favor, seleccione un mes del año en número: '))
```

```python
df['Mes'] = df['Fecha'].dt.month
df
```

```python
# Filtrar el DataFrame en base a la ciudad y el mes seleccionados
filtro = (df['Ciudad'] == ciudad_seleccionada) & (df['Mes'] == mes_seleccion)
resultados = df[filtro]

# Verificar si hay resultados
if not resultados.empty:
    # Crear una figura con dos subgráficos (uno al lado del otro)
    fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(12, 5))  # figsize ajusta el tamaño de la figura
    
    # Asignar los ejes a variables más claras
    ax1 = ax[0]  # Primer subgráfico
    ax2 = ax[1]  # Segundo subgráfico
    
    # Graficar temperaturas máximas en el primer subgráfico
    ax1.plot(resultados['Fecha'], resultados['Temperatura Maxima'], label='Temperatura Máxima', marker='o', color='red')
    ax1.set_title(f'Temperaturas Máximas en {ciudad_seleccionada} (Mes {mes_seleccion})')
    ax1.set_xlabel('Fecha')
    ax1.set_ylabel('Temperatura (°C)')
    ax1.legend()
    ax1.grid(True)
    ax1.tick_params(axis='x', rotation=45)
    
    # Graficar temperaturas mínimas en el segundo subgráfico
    ax2.plot(resultados['Fecha'], resultados['Temperatura Minima'], label='Temperatura Mínima', marker='o', color='blue')
    ax2.set_title(f'Temperaturas Mínimas en {ciudad_seleccionada} (Mes {mes_seleccion})')
    ax2.set_xlabel('Fecha')
    ax2.set_ylabel('Temperatura (°C)')
    ax2.legend()
    ax2.grid(True)
    ax2.tick_params(axis='x', rotation=45)
    
    # Mostrar el gráfico
    plt.show()
else:
    print('No hay resultados para la ciudad y el mes seleccionados.')
```

```python
# Filtrar el DataFrame en base a la ciudad y el mes seleccionados
filtro = (df['Ciudad'] == ciudad_seleccionada) & (df['Mes'] == mes_seleccion)
resultados = df[filtro]

# Verificar si hay resultados
if not resultados.empty:
    # Mostrar los resultados
    fig, ax = plt.subplots()
    
    # Graficar temperaturas máximas y mínimas
    ax.plot(resultados['Fecha'], resultados['Temperatura Maxima'], label='Temperatura Máxima', marker='o')
    ax.plot(resultados['Fecha'], resultados['Temperatura Minima'], label='Temperatura Mínima', marker='o')
    
    # Añadir título y etiquetas
    ax.set_title(f'Temperaturas en {ciudad_seleccionada} durante el mes {mes_seleccion}')
    ax.set_xlabel('Fecha')
    ax.set_ylabel('Temperatura (°C)')
    ax.legend()  # Mostrar leyenda
    ax.grid(True)  # Añadir cuadrícula

    plt.xticks(rotation=45)
    # Mostrar el gráfico
    plt.show()
else:
    print('No hay resultados para la ciudad y el mes seleccionados.')
```

```python
lista_ciudades = []

for c in df['Ciudad']:
    if c not in lista_ciudades:
        lista_ciudades.append(c)

lista_ciudades
```

```python
dict_meses = {1: 'Enero',
             2: 'Febrero',
             3: 'Marzo',
             4: 'Abril',
             5: 'Mayo',
             6: 'Junio',
             7: 'Julio',
             8: 'Agosto',
             9: 'Septiembre',
             10: 'Octubre',
             11: 'Noviembre',
             12: 'Diciembre'}
```

```python
def consultar_temperaturas():
    while True:
        # solicitar la ciudad
        print("\nCiudades disponibles: ", lista_ciudades)
        ciudad_elegida = input("Elija la ciudad de la lista: ").title()
        
        # Solicitar el mes
        mes_elegido = int(input("Elija un número de mes (por ej. Enero = 1): "))
        
        # validar los inputs
        if ciudad_elegida not in lista_ciudades or mes_elegido not in range(1, 13):
            print("Ciudad o mes no válidos. Inténtale nuevamente.")
            continue
        
        # Crear el datafreme filtrado
        datos_ciudad_mes = df[(df['Ciudad'] == ciudad_elegida) & (df['Fecha'].dt.month == mes_elegido)]
        
        # Graficamos las temperaturas del dataframe
        plt.figure(figsize=(10, 6))
        plt.plot(datos_ciudad_mes['Fecha'], datos_ciudad_mes['Temperatura Maxima'], label="Maxima", color="red")
        plt.plot(datos_ciudad_mes['Fecha'], datos_ciudad_mes['Temperatura Minima'], label="Minima", color="blue")
        plt.title(f"Temperaturas en {ciudad_elegida} durante el mes de {dict_meses[mes_elegido]}")
        plt.xlabel("Fecha")
        plt.ylabel("Temperaturas")
        plt.legend()
        plt.xticks(rotation=45)
        plt.show()
        
        # Preguntar al usuario si quiere continuar
        otra_consulta = input("¿Deseas continuar consultando? (s/n): ")
        if otra_consulta.lower() != 's':
            break
```

```python
consultar_temperaturas()
```
